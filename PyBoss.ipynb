```python
# import dependencies

import csv


# Create file path for the employee data csv

csvpath = 'employee_data.csv'

list_of_names = []  

# Skip header row in the csv



with open(csvpath, newline='') as csvfile:

    csvreader = csv.reader(csvfile,delimiter=',')
    if csv.Sniffer().has_header:
        next(csvreader)
        
#The Name column should be split into separate First Name and Last Name columns.

    for row in csvreader:
        str_names = str(row[1])
        list_of_names.append(str_names)
        
split_list_of_names = [i.split(' ') for i in list_of_names ]


first_name = []
last_name = []

for lst in split_list_of_names:
    first_name.append(lst[0])
    last_name.append(lst[1])
    
first_name
        
```

```python
#The DOB data should be re-written into MM/DD/YYYY format.

original_dob = []

with open(csvpath, newline='') as csvfile:

    csvreader = csv.reader(csvfile,delimiter=',')
    if csv.Sniffer().has_header:
        next(csvreader)

    for row in csvreader:
        str_dates = str(row[2])
        original_dob.append(str_dates)
        
original_dob
```

```python
import datetime

new_dob = []

for date in original_dob:
    new_dob.append(datetime.datetime.strptime(date, '%Y-%m-%d').strftime('%m/%d/%Y'))
    
new_dob
```

```python
#The SSN data should be re-written such that the first five numbers are hidden from view.

original_ssn = []

with open(csvpath, newline='') as csvfile:

    csvreader = csv.reader(csvfile,delimiter=',')
    if csv.Sniffer().has_header:
        next(csvreader)

    for row in csvreader:
        str_ssn = str(row[3])
        original_ssn.append(str_ssn)
        
original_ssn
```

```python
new_ssn = []
newer_ssn = []

for social in original_ssn:
    new_ssn.append(social[7:11])

for num in new_ssn:
    newer_ssn.append('***' + '-**-' + num)
    
newer_ssn
```

```python
#The State data should be re-written as simple two-letter abbreviations.

original_states = []

with open(csvpath, newline='') as csvfile:

    csvreader = csv.reader(csvfile,delimiter=',')
    if csv.Sniffer().has_header:
        next(csvreader)

    for row in csvreader:
        str_state = str(row[4])
        original_states.append(str_state)
        
original_states
```

```python
new_states = []

for state in original_states:
    new_states.append(us_state_abbrev[state])
    
new_states
```

```python
# Get list of employee Id's

Emp_ID = []

with open(csvpath, newline='') as csvfile:

    csvreader = csv.reader(csvfile,delimiter=',')
    if csv.Sniffer().has_header:
        next(csvreader)

    for row in csvreader:
        str_ID = str(row[0])
        Emp_ID.append(str_ID)
        
Emp_ID
```

```python
import pandas as pd

# Put our lists into dataframes

ID = pd.DataFrame(Emp_ID)
fname = pd.DataFrame(first_name)
lname = pd.DataFrame(last_name)
dob = pd.DataFrame(new_dob)
ssn = pd.DataFrame(newer_ssn)
states = pd.DataFrame(new_states)

# Merge and rename dataframes

results = pd.merge(ID,fname,left_index=True,right_index=True)
results2 = results.rename(columns={"0_x":"Emp ID","0_y":"First Name"})
results3 = pd.merge(results2,lname,left_index=True,right_index=True)
results4 = pd.merge(results3,dob,left_index=True,right_index=True)
results5 = results4.rename(columns={"0_x":"Last Name","0_y":"DOB"})
results6 = pd.merge(results5,ssn,left_index=True,right_index=True)
results7 = pd.merge(results6,states,left_index=True,right_index=True)
results8 = results7.rename(columns={"0_x":"SSN","0_y":"State"})
results9 = results8.set_index('Emp ID')

# Export dataframe to csv

results9.to_csv('new employee data.csv', encoding='utf-8')
```
